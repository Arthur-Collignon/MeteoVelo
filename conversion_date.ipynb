```python
import pandas as pd
```

```python
def convert_date_columns(df, date_columns, utc_bool=False):
    for col in date_columns:
        df[col] = pd.to_datetime(df[col], utc=utc_bool, errors='coerce', dayfirst=True)
        df[col] = df[col].dt.tz_localize('Europe/Paris')
    return df
```

```python
eau_df = pd.read_csv('./CSVs/eau.csv', sep=';')
temp_df = pd.read_csv('./CSVs/temperature.csv', sep=';')
```

```python
date_columns = ['horodate']
eau_converted = convert_date_columns(eau_df, date_columns)
temp_converted = convert_date_columns(temp_df, date_columns)
eau_converted
```

```python
# Conversion to csv
eau_converted.to_csv('./CSVs/eau_converted.csv', sep=';', index=False)
temp_converted.to_csv('./CSVs/temperature_converted.csv', sep=';', index=False)
```

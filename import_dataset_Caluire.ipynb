```python
import pandas as pd
```

```python
raw_data = pd.read_csv('./CSVs/Comptage_Plateau_Nord.csv', sep=';')
raw_data
```

```python
# replace NaN values with 0 and add count_b column to count column
raw_data = raw_data.fillna(0)
raw_data['count'] = raw_data['count'] + raw_data['count_b']
raw_data = raw_data.drop(columns=['count_b'])
raw_data
```

```python
# merge date column with debut and end time columns to create start_time and end_time columns
raw_data['start_time'] = raw_data['date'] + ' ' + raw_data['debut']
raw_data['end_time'] = raw_data['date'] + ' ' + raw_data['end']
raw_data
```

```python
raw_data.drop(columns=['date', 'debut', 'end'], inplace=True)
raw_data
```

```python
# Convert start_time and end_time columns to datetime format
raw_data['start_time'] = pd.to_datetime(raw_data['start_time'], utc=False, errors='coerce', dayfirst=True)
raw_data['end_time'] = pd.to_datetime(raw_data['end_time'], utc=False, errors='coerce', dayfirst=True)
raw_data['start_time'] = raw_data['start_time'].dt.tz_localize('Europe/Paris')
raw_data['end_time'] = raw_data['end_time'].dt.tz_localize('Europe/Paris')
raw_data
```

```python
# Export to CSV
raw_data.to_csv('./CSVs/counting_data_caluire.csv', sep=';', index=False)
```

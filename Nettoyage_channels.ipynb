```python
import pandas as pd
```

```python
channels_df = pd.read_csv('raw_data/channels-sites-comptage-metropole-lyon-corrected.csv', sep=';')
print(channels_df.head())

START_DATE = pd.to_datetime('2022-04-01', utc=True)
END_DATE = pd.to_datetime('2022-04-30', utc=True)

df_cleaned = channels_df.copy()
df_cleaned = df_cleaned.drop(columns=['started_at', 'last_updated_at'])
```

```python
# Remove lines where mobility_type is not BIKE
channels_df = channels_df[channels_df['mobility_type'] == 'BIKE']
print(f"Number of rows after filtering for BIKE: {len(channels_df)}")
channels_df
```

```python
# On vire les channels qui n'ont pas update après 2022-04-01
# Conversion de la colonne last_updated_at en datetime depuis le format string
df_cleaned['last_updated_at'] = pd.to_datetime(channels_df['last_updated_at'], utc=True, errors='coerce')
df_cleaned = df_cleaned[df_cleaned['last_updated_at'] >= START_DATE]
print(f"Number of rows after filtering for last update after 2023-04-01: {len(channels_df)}")
df_cleaned
```

```python
# Idem pour la colonne started_at mais cette fois on supprime les channels qui ont commencés après 2023-04-31
df_cleaned['started_at'] = pd.to_datetime(channels_df['started_at'], utc=True)
df_cleaned = df_cleaned[df_cleaned['started_at'] < END_DATE]
print(f"Number of rows after filtering for started before 2023-05-01: {len(channels_df)}")
df_cleaned
```

```python
# Convert dataframe to csv
df_cleaned.to_csv('./CSVs/channels.csv', sep=';', index=False)
```
